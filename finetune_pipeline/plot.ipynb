```python
import re 
import matplotlib.pyplot as plt
```

```python
epoch_agent_one = []
epoch_agent_two = []
epoch_agnet_three = []
train_perplexity_agent_one = []
train_perplexity_agent_two = []
train_perplexity_agent_three = []
train_epoch_loss_agent_one = []
train_epoch_loss_agent_two = []
train_epoch_loss_agent_three = []
```

```python
pattern = re.compile(r'Epoch (\d+): train_perplexity=([\d.]+), train_epoch_loss=([\d.]+)')
```

```python
files = ["finetune_output_3.1-8B-Instruct-100-epochs-first_agent.log", "finetune_output_3.1-8B-Instruct-100-epochs-second_agent.log", "finetune_output_3.1-8B-Instruct-100-epochs-third_agent.log"]
```

```python
for f in files:
    with open(f, 'r') as file:
        for line in file:
            match = pattern.search(line)
            if match:
                epoch = int(match.group(1))
                perplexity = float(match.group(2))
                loss = float(match.group(3))
                if "first" in f:
                    epoch_agent_one.append(epoch)
                    train_perplexity_agent_one.append(perplexity)
                    train_epoch_loss_agent_one.append(loss)
                elif "second" in f:
                    epoch_agent_two.append(epoch)
                    train_perplexity_agent_two.append(perplexity)
                    train_epoch_loss_agent_two.append(loss)
                else:
                    epoch_agnet_three.append(epoch)
                    train_perplexity_agent_three.append(perplexity)
                    train_epoch_loss_agent_three.append(loss)
```

```python
epoch_agent_one = list(set(epoch_agent_one))
epoch_agent_two = list(set(epoch_agent_two))
epoch_agnet_three = list(set(epoch_agnet_three))
```

```python
plt.plot(epoch_agent_one, train_perplexity_agent_one, label='Agent 1')
plt.xlabel('Epoch')
plt.ylabel('Perplexity')
plt.title("First Agent Perplexity against Epoch")
plt.legend()
plt.savefig("first_agent_perplexity.png")
plt.show()
```

```python
plt.plot(epoch_agent_two, train_perplexity_agent_two, label='Agent 2')
plt.xlabel('Epoch')
plt.ylabel('Perplexity')
plt.title("Second Agent Perplexity against Epoch")
plt.legend()
plt.savefig("second_agent_perplexity.png")
plt.show()
```

```python
plt.plot(epoch_agnet_three, train_perplexity_agent_three, label='Agent 3')
plt.xlabel('Epoch')
plt.ylabel('Perplexity')
plt.title("Third Agent Perplexity against Epoch")
plt.legend()
plt.savefig("third_agent_perplexity.png")
plt.show()
```

```python
plt.plot(epoch_agent_one, train_epoch_loss_agent_one, label='Agent 1')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title("First Agent Loss against Epoch")
plt.legend()
plt.savefig("first_agent_loss.png")
plt.show()
```

```python
plt.plot(epoch_agent_two, train_epoch_loss_agent_two, label='Agent 2')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title("Second Agent Loss against Epoch")
plt.savefig("second_agent_loss.png")    
plt.legend()
plt.show()
```

```python
plt.plot(epoch_agnet_three, train_epoch_loss_agent_three, label='Agent 3')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title("Third Agent Loss against Epoch")
plt.legend()
plt.savefig("third_agent_loss.png")
```

```python
orchestrator_epoch = []
orchestrator_perplexity = []
orchestrator_loss = []
```
